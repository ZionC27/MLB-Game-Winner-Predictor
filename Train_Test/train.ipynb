```python
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

# Assuming df is your DataFrame
X = df.drop(columns=['h_team_wins'], axis=1)
Y = df['h_team_wins']

# Verify data
print("X shape:", X.shape, "X dtypes:\n", X.dtypes)
print("Y shape:", Y.shape, "Y dtype:", Y.dtype)
print("Y unique values:", np.unique(Y))

# Handle missing values
X = X.fillna(X.mean())

# Split data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print("X_train shape:", X_train.shape)
print("Y_train shape:", Y_train.shape)
print("X_test shape:", X_test.shape)
print("Y_test shape:", Y_test.shape)

# Scale features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Initialize and fit LazyClassifier
clf = LazyClassifier(verbose=1, ignore_warnings=False, custom_metric=None)
models, predictions = clf.fit(X_train, X_test, Y_train, Y_test)

print("Models:\n", models)
print("Predictions:\n", predictions)
```

```python
# svc model
from sklearn.svm import SVC

# create a SVC classifier
svc = SVC(kernel='linear',probability=True)

# fit the classifier to the training data
svc.fit(X_train, Y_train)

# make predictions on the test data
y_pred = svc.predict(X_test)

# calculate the accuracy of the predictions
accuracy = svc.score(X_test, Y_test)
print("Accuracy:", accuracy)
```

```python
# using XGBoost model
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


X = df.drop(columns=['h_team_wins'], axis=1)
Y = df['h_team_wins']


# Handle missing values
X = X.fillna(X.mean())

# Split data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Initialize and fit XGBClassifier
model = XGBClassifier()
model.fit(X_train, Y_train)

# Make predictions
y_pred = model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy:", accuracy)    
```

```python
from sklearn.model_selection import GridSearchCV

# Define parameter grid for XGBClassifier
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0]
}

xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
grid_search = GridSearchCV(estimator=xgb, param_grid=param_grid, cv=3, scoring='accuracy', n_jobs=-1, verbose=2)

grid_search.fit(X_train, Y_train)

print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation accuracy: ", grid_search.best_score_)

# Evaluate on test set
best_xgb = grid_search.best_estimator_
y_pred_gs = best_xgb.predict(X_test)
test_accuracy = accuracy_score(Y_test, y_pred_gs)
print("Test set accuracy with best XGBoost: ", test_accuracy)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score


# Train Gradient Boosting model
gbc = GradientBoostingClassifier(random_state=42)
gbc.fit(X_train, Y_train)

# Predict on the same data (since no separate test set is provided here)
gbc_preds = gbc.predict(X_test)
gbc_accuracy = accuracy_score(Y_test, gbc_preds)

print("Gradient Boosting Accuracy", gbc_accuracy)
```

```python
from sklearn.neighbors import NearestCentroid
from sklearn.metrics import accuracy_score

# Create and fit the NearestCentroid classifier
nc = NearestCentroid()
nc.fit(X_train, Y_train)

# Predict on the test set
y_pred_nc = nc.predict(X_test)

# Calculate accuracy
accuracy_nc = accuracy_score(Y_test, y_pred_nc)
print("NearestCentroid Accuracy:", accuracy_nc)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

# Create and fit the neural network classifier
mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=300, random_state=42)
mlp.fit(X_train, Y_train)

# Predict on the test set
y_pred_nn = mlp.predict(X_test)

# Calculate accuracy
accuracy_nn = accuracy_score(Y_test, y_pred_nn)
print("Neural Network Accuracy:", accuracy_nn)

```

```python
gllc_df = pd.read_csv('gllc_cleaned.csv')
gldv_df = pd.read_csv('gldv_cleaned.csv')
glws_df = pd.read_csv('glws_cleaned.csv')
glwc_df = pd.read_csv('glwc_cleaned.csv')
```

```python
from sklearn.metrics import accuracy_score
import pandas as pd

# Prepare features and targets for each dataset
gllc_X = gllc_df.drop(columns=['h_team_wins'])
gllc_y = gllc_df['h_team_wins']

gldv_X = gldv_df.drop(columns=['h_team_wins'])
gldv_y = gldv_df['h_team_wins']

glws_X = glws_df.drop(columns=['h_team_wins'])
glws_y = glws_df['h_team_wins']

glwc_X = glwc_df.drop(columns=['h_team_wins'])
glwc_y = glwc_df['h_team_wins']

# Define datasets and their labels
datasets = {
    'gllc': (gllc_X, gllc_y),
    'gldv': (gldv_X, gldv_y),
    'glws': (glws_X, glws_y),
    'glwc': (glwc_X, glwc_y)
}

# Define models and their names
models = {
    'SVC': svc,
    'XGBoost': grid_search.best_estimator_,
    'NearestCentroid': nc,
    'MLP': mlp,
    'Gradient Boosting': gbc
}

# Dictionary to store accuracy scores for each model
model_accuracies = {model_name: [] for model_name in models}

# Function to generate predictions and return accuracy
def get_predictions_accuracy(model_name, model, X, y, dataset_name):
    preds = model.predict(X)
    accuracy = accuracy_score(y, preds)
    df = pd.DataFrame({'Actual': y, 'Predicted': preds})
    # Show all rows of the predictions DataFrame
    with pd.option_context('display.max_rows', None):
        print(f"\n{model_name} on {dataset_name} (Accuracy: {accuracy:.4f}):\n{df}")
    return accuracy

# Iterate over datasets and models to collect accuracies
for dataset_name, (X, y) in datasets.items():
    print(f"\n=== Predictions for {dataset_name} ===")
    for model_name, model in models.items():
        accuracy = get_predictions_accuracy(model_name, model, X, y, dataset_name)
        model_accuracies[model_name].append(accuracy)

# Calculate and display average accuracy for each model
print("\n=== Average Model Accuracies Across All Datasets ===")
avg_accuracies = {model_name: sum(accs) / len(accs) for model_name, accs in model_accuracies.items()}
for model_name, avg_accuracy in avg_accuracies.items():
    print(f"{model_name} Average Accuracy: {avg_accuracy:.4f}")
```

```python
import pandas as pd

# Load the new data for prediction
new_data = pd.read_csv('2025_06_09_LA_SDN_cleaned.csv')

# Define models
models = {
    'SVC': svc,
    'XGBoost': grid_search.best_estimator_,
    'NearestCentroid': nc,
    'MLP': mlp,
    'Gradient Boosting': gbc
}

# Generate predictions and confidence scores 
print("\n=== Predictions for 2025_06_09_LA_SDN ===")
results = []
for model_name, model in models.items():
    pred = model.predict(new_data)[0]  # Predict for the single row
    pred_label = "Home Team Wins" if pred == 1 else "Home Team Loses"
    
    # Get confidence score (probability of predicted class)
    if model_name == 'NearestCentroid':
        confidence = "N/A (model does not support probabilities)"
    else:
        try:
            probs = model.predict_proba(new_data)[0]  # Probabilities for [class 0, class 1]
            confidence = f"{probs[pred] * 100:.2f}%"  # Probability of predicted class
        except AttributeError:
            confidence = "N/A (probability not enabled)"
    
    results.append({
        'Model': model_name,
        'Prediction': pred_label,
        'Confidence': confidence
    })

# Display results in a clean table
print(pd.DataFrame(results).to_string(index=False))

# Majority vote for ensemble prediction
votes = [model.predict(new_data)[0] for model in models.values()]
majority_vote = 1 if sum(votes) > len(votes) / 2 else 0
majority_label = "Home Team Wins" if majority_vote == 1 else "Home Team Loses"
print(f"\nMajority Vote Prediction: {majority_label} ({sum(votes)}/{len(votes)} models predict home team wins)")
```
